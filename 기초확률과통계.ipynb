```python
# üìå ÌÜµÍ≥Ñ_codingÏó∞Ïäµ (3Ï£ºÏ∞®)
# 5Í∞ú Ïó∞Ïäµ ÏòàÏ†úÎ•º ÌïòÎÇòÏùò Colab ÎÖ∏Ìä∏Î∂ÅÏúºÎ°ú Ï†ïÎ¶¨

# Í≥µÌÜµ ÎùºÏù¥Î∏åÎü¨Î¶¨
import numpy as np
import matplotlib.pyplot as plt
import os
from matplotlib import font_manager as fm

# -----------------------------------
# Ïó∞Ïäµ01: Ï≤¥Ï§ë Ï∏°Ï†ï Îç∞Ïù¥ÌÑ∞ ÏÑ† Í∑∏ÎûòÌîÑ
# -----------------------------------
x = np.array([1, 2, 3, 4, 5, 6, 7])  # ÎÇ†Ïßú
y = np.array([78.4, 78.6, 78.5, 78.9, 78.2, 79.4, 78.6])  # Ï≤¥Ï§ë Îç∞Ïù¥ÌÑ∞

plt.plot(x, y, color="r")
plt.xlabel("Day")
plt.ylabel("Weight")
plt.title("Ïó∞Ïäµ01: ÏÑ† Í∑∏ÎûòÌîÑ")
plt.show()

# -----------------------------------
# Ïó∞Ïäµ02: Ï≤¥Ï§ë Ï∏°Ï†ï Îç∞Ïù¥ÌÑ∞ ÏÇ∞Ï†êÎèÑ
# -----------------------------------
plt.scatter(x, y, color="r")
plt.xlabel("Day")
plt.ylabel("Weight")
plt.title("Ïó∞Ïäµ02: ÏÇ∞Ï†êÎèÑ")
plt.show()

# -----------------------------------
# Ïó∞Ïäµ03: grid() Ï†ÅÏö© (ÏÑ†+ÎßàÏª§)
# -----------------------------------
plt.plot(x, y, color="r", marker="o")
plt.grid()
plt.xlabel("Day")
plt.ylabel("Weight")
plt.title("Ïó∞Ïäµ03: grid() Ï†ÅÏö©")
plt.show()

# -----------------------------------
# Ïó∞Ïäµ04: np.linspace + Îëê Í∞úÏùò Ìï®Ïàò Í∑∏ÎûòÌîÑ
# -----------------------------------
x2 = np.linspace(-5, 5, 300)  # -5ÏóêÏÑú 5ÍπåÏßÄ 300 Íµ¨Í∞Ñ
y1 = x2**2
y2 = (x2 - 2)**2

plt.plot(x2, y1, color="r", label="y1 = x^2")
plt.plot(x2, y2, color="k", linestyle="--", label="y2 = (x-2)^2")
plt.grid()
plt.legend()
plt.title("Ïó∞Ïäµ04: np.linspace ÏòàÏ†ú")
plt.show()

# -----------------------------------
# Ïó∞Ïäµ05: Í∑∏ÎûòÌîÑ Î™®Ïñë Íæ∏ÎØ∏Í∏∞ (ÌïúÍ∏Ä Ìè∞Ìä∏/Î≤îÎ°Ä/Ï∂ï/Í∑∏Î¶¨Îìú)
# -----------------------------------
# (Colab Ï†ÑÏö©) ÎÇòÎàîÍ≥†Îîï ÏÑ§Ïπò: ColabÏù¥Î©¥ Ï£ºÏÑù Ìï¥Ï†ú ÌõÑ Ïã§Ìñâ
# !apt-get update -qq
# !apt-get install -y fonts-nanum

font_path = "/usr/share/fonts/truetype/nanum/NanumGothic.ttf"
if os.path.exists(font_path):
    fm.fontManager.addfont(font_path)
    plt.rc("font", family=fm.FontProperties(fname=font_path).get_name())
plt.rcParams["axes.unicode_minus"] = False  # ÎßàÏù¥ÎÑàÏä§ Íπ®Ïßê Î∞©ÏßÄ

x3 = np.linspace(-5, 5)
y3 = 2 * x3
y4 = 5 * x3

plt.plot(x3, y3, label="y1 Í∞í", color="r")                     # Îπ®Í∞Ñ ÏÑ†
plt.plot(x3, y4, label="y2 Í∞í", linestyle="dashed", color="k")  # Í≤ÄÏ†ï Ï†êÏÑ†
plt.xlabel("X-Ï∂ï", size=14)
plt.ylabel("Y-Ï∂ï", size=14)
plt.title("Í∑∏ÎûòÌîÑ Ï†úÎ™©")
plt.grid()
plt.legend()
plt.show()
```
